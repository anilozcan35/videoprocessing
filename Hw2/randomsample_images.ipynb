```python
data_path = "/content/drive/MyDrive/Video Processing/Hw2/data/train"
```

```python
output_path = "/content/drive/MyDrive/Video Processing/Hw2/data"
```

```python
train_folder = os.path.join(output_path, 'train_set')
val_folder = os.path.join(output_path, 'eval')
```

```python
train_folder
```

```python
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
```

```python
imgs_list = [filename for filename in os.listdir() if os.path.splitext(filename)[-1] in image_extensions]
```

```python
len(imgs_list)
```

```python
random.shuffle(imgs_list)
```

```python
train_size = int(len(imgs_list) * 0.80)
val_size = int(len(imgs_list) * 0.20)
```

```python
for folder_path in [train_folder, val_folder]:
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
```

```python
os.path.exists(train_folder)
```

```python
for i, f in enumerate(imgs_list):
    if i < train_size:
        dest_folder = train_folder
    else:
        dest_folder = val_folder
    shutil.copy(os.path.join(data_path, f), os.path.join(dest_folder, f))
```

```python
os.rename(data_path, data_path.replace("train", "train_full") )
```

```python
os.rename(train_folder, train_folder.replace("train_set", "train") )
```
